```python
# Checking unique values
def uniques_dic(df):
    dic = {}
    
    for i in df:
        if len(df[i].unique()) < 50:
            dic[i] = list(df[i].unique())
            
    return dic
```

```python
# How many uniques per column
def uniques_len(df):
    dic_len = {}
    
    for i in df:
        dic_len[i] = len(df[i].unique())
        
    return dic_len
```

```python
import geopy.distance
import itertools
import folium
```

```python
def site_loc_tuples(df):
    locations = {}
    for (x, y, z) in zip(df['Am Id'], df['Lat'], df['Lon']):
        locations[str(x)] = (y, z)
    return locations
```

```python
loc = site_loc_tuples(sites)
```

```python
def get_km(df, x):
    '''From a dataframe df calculate the distance in kms from a selected site to every other site in the dataframe.
    
    Parameters:       
            df = Dataframe you will use
            x = index of selected/current site
    '''
    
    distances = []
    current_site = str(df['Am Id'][x])
    for i in df.index:
        other_site = str(df['Am Id'][i])
        distances.append(geopy.distance.distance(loc[current_site], loc[other_site]))
        
    dist_df = pd.DataFrame(distances, columns=[current_site], index=df['Am Id'])
                         
    return dist_df
```

```python
get_km(sites, 0)
```

```python
def get_nearby(df, x, k):
    '''Get all nearby sites from a selected site from a dataframe df that fall within a km radius of x
    Parameters:
            df = Dataframe in use
            x = index of selected/current site
            k = Maximum distance(km) defining nearby sites'''
    
    distances = []
    nearby = []
    current_site = str(df['Am Id'][x])
    
    for i in df.index:
        other_site = str(df['Am Id'][i])
        if geopy.distance.distance(loc[current_site], loc[other_site]) < k:
            nearby.append(other_site)
            distances.append(geopy.distance.distance(loc[current_site], loc[other_site]))
            
    addresses = [df[['Am Id', 'Lat', 'Lon', 'Address']][df['Am Id']==i] for i in df['Am Id'] if str(i) in nearby]
    address_df = addresses[0]
    for i in range(1, len(addresses)):
        address_df = pd.concat([address_df, addresses[i]], axis=0)
        
    address_df['Distance'] = distances
    
    return address_df
```

```python
get_nearby(sites, 0, 1)
```

```python
def plot_nearby_sites(ID, radius): # do not exceed 1 km
    
    table = get_nearby(sites, ID, radius)
    first_point = table.iloc[0, :]
    
    maps = folium.Map(location = (first_point['Lat'], first_point['Lon']), 
                      tiles='Stamen Terrain', zoom_start=14)
    
    for i in range(0, len(table)):
        folium.Marker((table.iloc[i]['Lat'], table.iloc[i]['Lon']), popup= 
                     table.iloc[i]['Address']).add_to(maps)
    
    return display(maps)
```
