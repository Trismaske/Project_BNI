# Functional Outputs

Be sure to uncomment the geopy installation cell and run it to install the geopy library for the distance calculation

```python
def get_address():
    address = input("Enter current site:")
    return address
        
def get_distance():
    radius = float(input("Enter distance:"))   
    return radius

def get_id():
    site = int(input("Enter current site:"))
    return site
```

```python
def nearby_ids(df):
    '''Generates a table with the sites within radius of the selected site'''
    
    df = df[['Am Id','Lat', 'Lon', 'Address']]
    
    current_site = df[df['Am Id'] == get_id()] # Filter only the features of the selected site in a dataframe
        
    # Create latitude and longitude variables for the selected site
    lat = current_site['Lat'].values
    lon = current_site['Lon'].values
    
    distances = [] # Create an empty list for distances
    
    # A for loop populating the empty list
    for i in df.index:                                                            
        distances.append(geopy.distance.distance((lat, lon), (df['Lat'][i], df['Lon'][i])))
    
    # Create a new Datarame with the distances list, and add other features
    table = pd.DataFrame(distances, columns=['distance (km)'])
    table['Am Id'] = df['Am Id']
    table['Address'] = df['Address']
    table['Lat'] = df['Lat']
    table['Lon'] = df['Lon']
    table = table[['Am Id','Address','distance (km)','Lat','Lon']]
    table['distance (km)'] = table['distance (km)'].astype(str).str[:-3].astype(float)
    
    table = table[table['distance (km)'] <= get_distance()] # Filter table based on distance
    
    return table
```

```python
nearby_ids(sites)
```

```python
def nearby_addresses(df):
    '''Generates a table with the sites within radius of the selected site'''
    
    df = df[['Am Id','Lat', 'Lon', 'Address']]
    
    current_site = df[df['Address'] == get_address()] # Filter only the features of the selected site in a dataframe
    
    # Create latitude and longitude variables for the selected site
    lat = current_site['Lat'].values
    lon = current_site['Lon'].values

    distances = []  # Create an empty list for distances
    
    # A for loop populating the empty list
    for i in df.index:                                                            
        distances.append(geopy.distance.distance((lat, lon), (df['Lat'][i], df['Lon'][i])))
        
    # Create a new Datarame with the distances list, and add other features
    table = pd.DataFrame(distances, columns=['distance (km)'])
    table['Am Id'] = df['Am Id']
    table['Address'] = df['Address']
    table['Lat'] = df['Lat']
    table['Lon'] = df['Lon']
    table = table[['Am Id','Address','distance (km)','Lat','Lon']]
    table['distance (km)'] = table['distance (km)'].astype(str).str[:-3].astype(float)
    
    table = table[table['distance (km)'] <= get_distance()] # Filter table based on distance
    
    return table
```

```python
handler = []
```

```python
def on_button_clicked(b):
    if b.description == 'Choose ID input':
        global handler 
        handler.append('ID')
        return print(handler)
    
    elif b.description == 'Choose Address input':
        handler.append('Address')
        return print(handler)
```

```python
ID = Button(description='Choose ID input')
Address = Button(description='Choose Address input')  
```

```python
ID.on_click(on_button_clicked)
Address.on_click(on_button_clicked)
```

```python
AppLayout(header=ID,
          footer=Address)
```

```python
def handle():
    if handler[-1] == 'ID':
        return nearby_ids(sites)
    else:
        return nearby_addresses(sites)
```

```python
handle()
```

```python
def plot_nearby_sites(): # do not exceed 1 km
    '''Generates a map with the sites within the radius of the selected site'''
    
    # Generate table
    table = handle()
    first_point = table.iloc[0, :]
    
    # Generate map
    maps = folium.Map(location = (first_point['Lat'], first_point['Lon']), 
                      tiles='OpenStreetMap', zoom_start=14)
    
    # Add sites with markers
    for i in range(0, len(table)):
        folium.Marker((table.iloc[i]['Lat'], table.iloc[i]['Lon']), popup= 
                    table.iloc[i]['Address']).add_to(maps)
    
    
    return display(maps)
```

```python
print(plot_nearby_sites())
```
